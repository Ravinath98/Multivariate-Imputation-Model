```python
#import the required libraries
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
```

```python
#read the dataset
df=pd.read_csv('heart_disease_data.csv')
```

```python
#head values
df.head()
```

```python
#dataset information
df.info()
```

```python
#filter only 1 or more nan values rows
df1  = df[df.isna().any(axis=1)]
print(df1.shape)
df1.head()
```

```python
#dataframe which not containing null values
df2=df.dropna()
df2
```

```python
#replace null values to 0 and not null values to 1(for the help for grouping in the next codes)
df3=df1.notnull().astype("int")
df3.head()
```

```python
#grouping the dataframe based on null value quantity and position
dfGrouped=df3.groupby(list(df3.columns))
```

```python
data_frame = pd.DataFrame() #create empty dataframe
for k in dfGrouped.groups: 
    list_of_indexes = list(dfGrouped.get_group(k).index) #get the indexes of corresponding group(same # of null values with same position)
    tempdf = df.iloc[list_of_indexes, :] #get the corresponding dataframe part from the main dataframe(using indexes)
    nullcols=list(tempdf.columns[tempdf.isnull().any()]) #get null cols in the sub dataframe
    x=df2.drop(nullcols, axis = 1) #define x from the not null dataframe
    y=df2[nullcols] #define y from the not null dataframe
    LR=LinearRegression() #create the model
    LR.fit(x,y) #fit the model
    pred=LR.predict(tempdf.drop(nullcols, axis = 1)) #predict from the model
    #replacing missing values with prediction values...
    for i in range(len(nullcols)): 
        tempdf[nullcols[i]]=pred[:,i]
   

    data_frame=pd.concat([tempdf,data_frame]) #concat each dataframe, after predict and replace values with nan
finaldf=pd.concat([df2,data_frame]) #concat the new null values filled dataframe and the dataframe which not contained the null values at the beginning
finaldf=finaldf.sort_index() #final dataframe after filling the missing values
```
